### Loading data

```python
train = pd.read_excel(os.path.join('data', 'xlsx', 'Data_Train.xlsx'))
test = pd.read_excel(os.path.join('data', 'xlsx', "Data_Test.xlsx"))
data = pd.concat((train, test))
```

```python
data.head()
```

### Data Transformations

##### Features: 
 - Title: The title of the book 
 - Author: The author(s) of the book. 
 - Edition: The edition of the book eg (Paperback,– Import, 26 Apr 2018) 
 - Reviews: The customer reviews about the book 
 - Ratings: The customer ratings of the book 
 - Synopsis: The synopsis of the book 
 - Genre: The genre the book belongs to 
 - BookCategory: The department the book is usually available at. 
 - Price: The price of the book (Target variable)


##### No. of Records

```python
len_train, len_test = len(train), len(test)
print("# train:", len_train)
print("# test:", len_test)
```

###### Checking missing values

```python
data.isna().sum()
```

##### Feature Engg: Reviews
Getting integer value

```python
data['Reviews'] = data['Reviews'].apply(lambda x: float(x.split()[0]))
```

```python
# data.head()
```

##### Feature Engg: Ratings
Getting no. of rating (int value)

```python
data['Ratings'] = data['Ratings'].apply(lambda x: (x.split()[0].replace(',',''))).astype(int)
```

```python
data.head()
```

##### Feature Engg: Transforming following columns to lower
- Title
- Author
- Edition
- Synopsis
- Genre
- BookCategory

```python
for col in ['Title','Author','Edition','Synopsis','Genre','BookCategory']:
    data[col] = data[col].str.lower()
```

```python
data.head()
```

##### Feature Engg: Genre
Cleanup and removing `(books)`, `Textbooks`

```python
data['Genre'] = data['Genre'].apply(lambda x: x.strip('(books)').strip('Textbooks'))
```

```python
# data.head()
```

##### Feature Engg: Extracting `Edition_Year` of publish from the `Edition` column

```python
data['Edition_Year'] = data['Edition'].apply(lambda x: x.split()[-1] if x.split()[-1].isdigit() else 'na')
```

```python
# data.head()
```

##### Feature Engg: Extracting `Binding` of publish from the `Edition` column

```python
data['Binding'] = data['Edition'].apply(lambda x: np.where('paperback' in x, 1, 0))
```

```python
# data.head()
```

##### Feature Engg: Merging `Synopsis` and `Title`

```python
data['Synopsis'] = data['Synopsis'] + " " + data['Title']
```

```python
data.head()
```

### Preparing data for model training

```python
train, test = data[:len_train], data[len_train:]
y_train = train.pop('Price')
test.pop('Price');

X_train = train
X_test = test
```

### `LGBM Regressor`: Model parameters for gridsearch

LightGBM is a gradient boosting framework that uses tree based learning algorithms. It is designed to be distributed and efficient with the following advantages:
- Faster training speed and higher efficiency.
- Lower memory usage.
- Better accuracy.
- Support of parallel and GPU learning.
- Capable of handling large-scale data.
- LightGBM can use categorical features directly (without one-hot encoding). hows about 8x speed-up compared with one-hot encoding.

```python
# FT for datasets
data_ft = FunctionTransformer(lambda x: x[['Author', 'Reviews', 'Ratings', 'Genre', 'BookCategory', 'Edition_Year', 'Binding']], validate = False)
pipeline_ft = Pipeline([('selector', data_ft),])
```

```python
params = {
    'clf__n_estimators' : [100, 200],
    #'clf__max_depth' : [3,4,5,6,7],
    'clf__learning_rate': [0.01, 0.1, 0.01],
    'clf__reg_lambda': list(np.arange(0.1, 0.9, 0.1)),
    #'clf__colsample_bytree' : list(np.arange(0.1,0.8,0.1)),
    'clf__importance_type': ['gain', 'weight', 'cover', 'total_gain', 'total_cover'],
    #'clf__booster': ['gbtree', 'gblinear', 'dart']
}
```

```python
pipeline = Pipeline([('numeric', pipeline_ft), ('clf', LGBMRegressor(verbosity = 1, objective = 'regression'))])
```

##### Custom Storing Method

```python
def get_score(y_val, y_pred):
    return np.sqrt(mean_squared_log_error(y_pred, y_val))
    
criteria = make_scorer(get_score, greater_is_better = False)
grid = RandomizedSearchCV(pipeline, param_distributions = params, n_iter = 15, cv = 5, scoring = criteria, verbose = 0);
grid.fit(X_train, y_train);
```

```python
print(grid.best_params_)
print('Best Score: ', grid.best_score_)
```

##### Make predictions on train data

```python
y_pred_train = grid.predict(X_train)
print('Mean Squared Error: ', mean_squared_error(y_pred_train, y_train))

print('Validation set score: ', 1 - np.sqrt(np.square(np.log10(y_pred_train +1) - np.log10(y_train +1)).mean()))
```

```python
OUTPUT = pd.DataFrame(grid.predict(test), columns=['Price'])

OUTPUT.to_csv('submissions.csv',index=None)
```
