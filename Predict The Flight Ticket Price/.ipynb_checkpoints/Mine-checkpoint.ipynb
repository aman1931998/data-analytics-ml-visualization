## Predict The Flight Ticket Price Hackathon

Flight ticket prices can be something hard to guess, today we might see a price, check out the price of the same flight tomorrow, it will be a different story. We might have often heard travellers saying that flight ticket prices are so unpredictable. Huh! Here we take on the challenge! As data scientists, we are gonna prove that given the right data anything can be predicted. Here you will be provided with prices of flight tickets for various airlines between the months of March and June of 2019 and between various cities.
Size of training set: 10683 records 
Size of test set: 2671 records 
FEATURES: 

- Airline: The name of the airline.
- Date_of_Journey: The date of the journey.
- Source: The source from which the service begins.
- Destination: The destination where the service ends.
- Route: The route taken by the flight to reach the destination.
- Dep_Time: The time when the journey starts from the source.
- Arrival_Time: Time of arrival at the destination.
- Duration: Total duration of the flight.
- Total_Stops: Total stops between the source and destination.
- Additional_Info: Additional information about the flight.
- Price: The price of the ticket - TARGET

### Goal
Creating a machine learning for predicting flight ticket price with high accuracy.

### Feature Engineering

###### Creating a travel class feature to distinguish between economy, Business and premium economy

```python
data['Travel_Class'] = data['Airline'].map(
    {
        'IndiGo': 'Economy', 
        'GoAir': 'Economy', 
        'Vistara': 'Economy', 
        'Vistara Premium economy': 'Premium Economy', 
        'Air Asia': 'Economy', 
        'Trujet': 'Economy', 
        'Jet Airways': 'Economy', 
        'SpiceJet': 'Economy', 
        'Jet Airways Business': 'Business', 
        'Air India': 'Economy', 
        'Multiple carriers': 'Economy', 
        'Multiple carriers Premium economy': 'Premium Economy'
    }
)
data['Travel_Class'] = data['Travel_Class'].astype('category')
```

###### Creating a market share column to map each airlines market share in india

```python
data['Market Share'] = data['Airline'].map(
    {
        'IndiGo': 43, 
        'GoAir': 9, 
        'Vistara': 4,
        'Vistara Premium economy': 2,
        'Air Asia': 5,
        'Trujet': 0.5,
        'Jet Airways': 12,
        'SpiceJet': 13,
        'Jet Airways Business': 8,
        'Air India': 11,
        'Multiple carriers': 2,
        'Multiple carriers Premium economy': 2
    }
)
data['Market Share'] = data['Market Share'].astype('int')
```

###### Creating a airline ranking feature to rank the airline on the basis of ranking given to each airline in 2019

```python
data['Airline Ranking']=data['Airline'].map(
    {
        'IndiGo':1,
        'Air India':2,
        'Jet Airways':2,
        'SpiceJet':3,
        'GoAir':4,
        'Vistara':5,
        'Air Asia':6,
        'Trujet':9,
        'Vistara Premium economy':5,
        'Multiple carriers Premium economy':6,
        'Multiple carriers':8,
        'Jet Airways Business':2
    }
)
data['Airline Ranking'] = data['Airline Ranking'].astype('int')
```

##### Converting `Total_Stops` to int

```python
data["Total_Stops"] = data["Total_Stops"].replace("non-stop","0 stops")
data["Total_Stops"] = data["Total_Stops"].str.split(" ").str[0]

data['Total_Stops'] = data['Total_Stops'].astype('int')
```

##### Checking `Additional_Info` column

```python
data["Additional_Info"].unique()
```

##### Converting Duration to int value (minutes)

```python
ddata = list(map(lambda x: x.split(), data['Duration'].values))
duration_data = []
for i in ddata:
    length = len(i)
    val = int(i[0][:-1]) * 60
    if(length == 2):
        val+= int(i[1][:-1])
    duration_data.append(val)
data['Duration'] = duration_data

del ddata, duration_data, i, val, length
```

##### Getting distance travelled for each flight via web scraping on prokerela.com

```python
####################################Path#######################################
route = [(list(map(lambda x:x.strip(), i.split('→')))) for i in data['Route'].values]
x = set()
for i in route: 
    for j in i: 
        x.add(j)
airports = list(x)
airports_combinations = []
for i in combinations(airports, 2):
    airports_combinations.append(i)

del x, airports
```

```python
try:
    with open('data/distances.pkl', 'rb') as file:
        distances = pkl.load(file)
except:
    c, distances = 0, []
    for i, j in airports_combinations:
        url = 'https://www.prokerala.com/travel/airports/distance/from-' + i.lower() + '/to-' + j.lower() + '/'
        print(url)
        yelp_r = requests.get(url)
        yelp_soup = BeautifulSoup(yelp_r.text, 'html.parser')
        distances.append(str(yelp_soup.findAll(name = 'h2')[0]).split()[3])
        c+=1
        print(c)
    with open('data/distances.pkl', 'wb') as file:
        pkl.dump(distances, file)

```

```python
distance, c = {}, 0
for i, j in airports_combinations:
    distance[i+j] = float(distances[c])
    distance[j+i] = float(distances[c])
    c+=1

distance_a2a = []
for i in route:
    
    length, dist = len(i)-1, 0
    for j in range(length):
        dist += distance[i[j] + i[j+1]]
    distance_a2a.append(dist)

data['Distance'] = np.array(distance_a2a, dtype = np.float32).reshape((len_train + len_test, 1))

# del airports_combinations, c, dist, distance, distance_a2a, distances, i, j, length, route, url, yelp_r, yelp_soup
###############################################################################
```

##### Used TF-IDF to get the route feature in a processed way

```python
data['Route'] = data['Route'].apply(lambda x: ' '.join(x.split(' → ')))

from sklearn.feature_extraction.text import TfidfVectorizer
tfv = TfidfVectorizer(ngram_range=(1, 1), lowercase=False)
data_route = tfv.fit_transform(data['Route']).toarray()
data_route = pd.DataFrame(data = data_route, columns = tfv.get_feature_names())
data_route.index = data.index
```

```python
data = pd.concat((data, data_route), axis = 1)
data = data.drop(columns = ['Route'])
```

```python
data['Airline'] = data['Airline'].astype('category')
data['Source'] = data['Source'].astype('category')
data['Destination'] = data['Destination'].astype('category')
data['Additional_Info'] = data['Additional_Info'].astype('category')
```

##### Getting Hour of departure
Using criteria when `minute` is >= 50, we add round off the `hour`

```python
data['Dep_Hour'] = data['Dep_Time'].str.split(":").apply(lambda x: x[0]).astype('int') + \
                        (data['Dep_Time'].str.split(":").apply(lambda x: x[1]).astype('int') > 50).astype('int')
```

```python
data = data.drop(columns = ['Dep_Time', 'Arrival_Time', 'Date_of_Journey'])
```
